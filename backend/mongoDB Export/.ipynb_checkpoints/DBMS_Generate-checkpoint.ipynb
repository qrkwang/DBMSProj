```python
tempUser= pd.read_csv('username_password.csv', sep=';')
tempUser
```

```python
tempUserOthers= pd.read_csv('user_others.csv', sep=',')
tempUserOthers
```

```python
customerDF=pd.DataFrame(columns=['customerid'])

customerDF['name']=tempUserOthers['first_name']+" "+tempUserOthers['last_name']
customerDF['customerid']=customerDF.index+1
contact=np.random.randint(80000000,99999999, size=len(pd.DataFrame(customerDF['customerid'])))
customerDF['contactNo']=contact.astype(str)
customerDF['address']= tempUserOthers['address']


customerDF['username']=np.random.choice(tempUser['Username'], len(customerDF))+customerDF.index.astype(str)
customerDF['password']=np.random.choice(tempUser['One-time password'], len(customerDF))+customerDF.index.astype(str)
customerDF
```

```python
#MongoDB insert
#customer collection
company= db["customer"]
data_dict = customerDF.head(10).to_dict(orient='records')
company.insert_many(data_dict)
```
